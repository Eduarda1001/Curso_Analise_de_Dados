Estatística Básica

```python
#A biblioteca Numpy permite usar funções estatísticas
import numpy as np
```

```python
lista_idades = [26,30,32,22,26,35,40,20,43,31,23,100]
```

```python
#Soma das idades
np.sum(lista_idades)
```

```python
#Média das idades
#A função "mean" calcula a média aritmética
media = np.mean(lista_idades)
print("Média Aritmética =", media)
```

```python
#A função "sort" bota os dados em ordem crescente
lista_idades.sort()
lista_idades
```

```python
#A função "median" calcula a mediana
mediana = np.median(lista_idades)
print("Mediana =", mediana)
```
